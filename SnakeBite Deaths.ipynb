```python
# Import Libraries of python.
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Scrapped a WEBSITE elifescience.org.
url='https://elifesciences.org/articles/54076'
```

```python
abhi=requests.get(url)
abhi
```

```python
# Access the html file using BeautifulSoup.
soup=BeautifulSoup(abhi.text)
soup
```

```python
# find the table in above html text.
table3 = soup.find('div', id="table3")
table = table3.find_all('td')
```

```python
table
```

```python
# Converted the HTL values in List.
content = []
for i in table:
    content.append(i.text)
content
```

```python
for i,c in enumerate(content):
    print(i,c)    
```

```python
#append the States in the empty list According to the Scrapped Data
states = []

for i in range(len(content)):
    if not content[i].isnumeric() and not content[i].replace('.','').isnumeric() and content[i] != '':
        states.append(content[i])
```

```python
print(states)
print(len(states))
```

```python
deaths=[]

for i in range(len(content)):
    if content[i].isnumeric():
        deaths.append(content[i])
```

```python
print(deaths)
len(deaths)
```

```python
# append the total estimated death rate over every State.
total = []
for i in range(len(content)):
    if not content[i].isnumeric():
        if content[i].replace('.','').isnumeric():
            total.append(content[i])
```

```python
print(total)
len(total)
```

```python
for i,t in enumerate(total):
    print(i,t)
```

```python
#Putting the avg death rates according to years.
avg_year1 = []
avg_year2 = []
avg_year3 = []
avg_year = []

i = 0
while i < len(total):
    avg_year1.append(total[i])
    i += 4
```

```python
i = 1
while i < len(total):
    avg_year2.append(total[i])
    i += 4
```

```python
i = 2
while i < len(total):
    avg_year3.append(total[i])
    i += 4
```

```python
i = 3
while i < len(total):
    avg_year.append(total[i])
    i += 4
```

```python
#checking the length of the columns are balanced or not.
len(avg_year1)
```

```python
len(avg_year2)
```

```python
len(avg_year3)
```

```python
len(avg_year)
```

```python
#Put the Scrapped Data into Dictionary beacause we want key and values
data = {
    'State': states,
    'Study deaths in MDS': deaths,
    'Average death 2001-2004': avg_year1,
    'Average death 2005-2009': avg_year2,
    'Average death 2010-2014': avg_year3,
    'Estimated deaths for 2001-14': avg_year,
}
```

```python
data
```

```python
#  Converted the Scrapped Data into a DataFrame.
snakebite = pd.DataFrame(data)
snakebite
```

```python
# Save the above dataframe into CSV(Comma Separated Value) file. 
snakebite.to_csv('SnakeBite.csv',index=False)
```
